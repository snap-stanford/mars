```python
#color differently
X_tsne = myTSNE.X_tsne
colors_list = ['#e41a1c','#a65628','#4daf4a','#ff7f00','#FFD700','#377eb8','#f781bf','#984ea3','#999999','#00CED1']
l2c = {}
idx_color = 0
plt.figure()
plt.figure(figsize=(10,10))
for idx, val in enumerate(X_tsne):
    x = val
    label = mars_labels[idx]
    if label not in l2c:
        l2c[label] = colors_list[idx_color]
        idx_color +=1
    curr_color = l2c[label]
    if x[0]>8 and x[1]<-30: # color differently Lim1 cluster
        curr_color = colors_list[-1]
    plt.scatter(x[0], x[1], marker='o', facecolors=curr_color,edgecolors=curr_color, s=50, label=label)
plt.xlabel('tSNE 1')
plt.ylabel('tSNE 2')
plt.savefig('/Users/maria/Desktop/fig1.pdf', dpi=None, format='pdf')
plt.tight_layout()
```

# Plot clusters

```python
import hdbscan
clusterer = hdbscan.HDBSCAN(min_cluster_size=5, min_samples=5).fit(myTSNE.X_tsne)
labels_HDBSCAN = clusterer.labels_
print("Clusters", max(labels_HDBSCAN)+1)
```

```python
# Plot TSNE colored by cluster assignment with labels
fig, ax = plt.subplots(1, 1, figsize=(10,10))
myTSNE.plot(fig, ax, colorMode="custom", colorBy=labels_HDBSCAN, cmap="nipy_spectral")
for label in pd.Series(labels_HDBSCAN).value_counts().index:
    if label == -1: continue
    #x_bar = np.mean(myTSNE.X_tsne[labels_HDBSCAN == label, 0])
    #y_bar = np.mean(myTSNE.X_tsne[labels_HDBSCAN == label, 1])
    #ax.text(x_bar, y_bar, label, size=18, color = 'r')

# fig.savefig("../figure/knot_dev/knot_24h_48_adult_hdbscan_clusters.pdf")
```
